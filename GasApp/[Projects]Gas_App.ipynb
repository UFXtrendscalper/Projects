# Gas Price Comparison Tool

The Gas Price Comparison Tool is a handy application that allows you to compare the cost of filling your gas tank between the United States and Canada. By inputting your gas tank size in liters, the tool calculates the cost of filling your tank in both countries based on the prevailing gas prices. It then determines which option, either the USA or Canada, offers a cheaper price for refueling. The tool also provides an estimate of the potential savings you can enjoy by choosing the more affordable option. With this information, you can make an informed decision about where to purchase gas and potentially save money on your fuel expenses.

Notes:
    The average vehicle gas tank can carry between 45 to 65 liters of fuel.

Gasbuddy:: https://www.gasbuddy.com/gaspricemap?fuel=1&z=14&lat=49.01128947922177&lng=-122.26687927501227

```python
# Gas Price Comparison Calculator

# Define gas prices and conversion rates
usd_gas_price = 4.29  # USD per gallon
cad_gas_price = 1.85  # CAD per litre
one_usd = 1.34  # USDCAD conversion rate
one_gallon = 3.78541  # liters per gallon

# Get user input for gas tank size in liters
tank_size_liters = float(input("Enter your gas tank size in liters: "))

# Calculate the cost of filling the tank in the USA
tank_size_gallons = tank_size_liters / one_gallon
usd_cost = tank_size_gallons * usd_gas_price * one_usd

# Calculate the cost of filling the tank in Canada
cad_cost = tank_size_liters * cad_gas_price

# Determine the cheaper option
if usd_cost < cad_cost:
    cheaper_option = "USA"
    savings = cad_cost - usd_cost
else:
    cheaper_option = "Canada"
    savings = usd_cost - cad_cost

# Print the result
print("Gas Price Comparison Results:")
print("=====================================")
print("It is cheaper to buy gas in", cheaper_option)
print("You would save $", round(savings, 2))
print('USD cost:', round(usd_cost, 2))
print('CAD cost:', round(cad_cost, 2))
```
